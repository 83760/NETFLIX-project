```python
# Extract the year from the date_added column
flix['year_added'] = pd.to_datetime(flix['date_added']).dt.year

# Filter for TV shows
tv_shows = flix[flix['type'] == 'TV Show']

# Count the number of TV shows by year
tv_shows_by_year = tv_shows['year_added'].value_counts().sort_index()

# Plotting the distribution of TV shows by year
plt.figure(figsize=(15, 6))
sns.lineplot(x=tv_shows_by_year.index, y=tv_shows_by_year.values, color='darkblue')
plt.title('Number of TV Shows Released per Year', fontsize =21)
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
#INSIGHTS :- Based on above data obtained, sudden raise in trend from 2015 onwards.
```

```python
# Extract the year from the date_added column
flix['month_added'] = pd.to_datetime(flix['date_added']).dt.month

# Filter for TV shows
tv_shows = flix[flix['type'] == 'TV Show']

# Count the number of TV shows by year
tv_shows_by_year = tv_shows['month_added'].value_counts().sort_index()

# Plotting the distribution of TV shows by year
plt.figure(figsize=(15, 6))
sns.lineplot(x=tv_shows_by_year.index, y=tv_shows_by_year.values, color='darkgreen')
plt.title('Number of TV Shows Released per Month', fontsize = 21)
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
#INSIGHTS :- Based on above data obtained, sudden raise in trend in the month of july and december.
```

```python
# Filter for recent years (e.g., 2018 onwards)
recent_years = flix[flix['release_year'] >= 2018]

# Count the number of TV shows and movies in each year
content_counts = recent_years['type'].value_counts()

# Plotting the focus of Netflix on TV shows versus movies
plt.figure(figsize=(8, 6))
plt.bar(content_counts.index, content_counts.values, color=['darkblue', 'darkorange'])
plt.title('Focus of Netflix on TV Shows vs Movies (Recent Years Overall)', fontsize = 15)
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
# Filter for recent years (e.g., 2018 onwards)
recent_years = flix[flix['release_year'] >= 2018]

# Group by year and content type, and count the number of occurrences
yearly_counts = recent_years.groupby(['release_year', 'type']).size().unstack()

# Plotting the year-wise focus of Netflix on TV shows versus movies
fig, ax = plt.subplots(figsize=(15, 6))
yearly_counts.plot(kind='bar', stacked=True, ax = ax)
plt.title('Focus of Netflix on TV Shows vs Movies (Yearwise)', fontsize = 21)
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Content Type')
plt.show()
```

```python
#INSIGHTS :- Based on above data obtained, It is seen that in 2021 contents from movies and tv shows are almost same.
#And also it is same in recent years.
```

```python
# Filter the data where cast_NA is not equal to flix['cast'] and director_NA is not equal to flix['director']
filtered_data = flix[(flix['cast'] != 'cast_NA') & (flix['director'] != 'director_NA')]

# Extracting the actors and directors columns
actors = filtered_data['cast']
directors = filtered_data['director']

# Counting the occurrences of actors and directors for each content type
actor_counts = actors.value_counts()
director_counts = directors.value_counts()

# Plotting the overall top 10 actors
plt.figure(figsize=(14, 6))
actor_counts.head(10).plot(kind='bar', color='purple', alpha=0.7)
plt.title('Overall Top 10 Actors', fontsize = 21)
plt.xlabel('Actor')
plt.ylabel('Count')
plt.show()

# Plotting the overall top 10 directors
plt.figure(figsize=(14, 6))
director_counts.head(10).plot(kind='bar', color='blue', alpha=0.7)
plt.title('Overall Top 10 Directors', fontsize = 21)
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
# Filter the data where cast_NA is not equal to flix['cast']
filtered_data = flix[(flix['cast'] != 'cast_NA')]

# Extracting the actors columns
actors = filtered_data['cast']


# Counting the occurrences of actors and directors for each content type
actor_counts_tv = actors[flix['type'] == 'TV Show'].value_counts()
actor_counts_movie = actors[flix['type'] == 'Movie'].value_counts()

# Plotting the top 10 actors in TV shows
plt.figure(figsize=(14, 6))
actor_counts_tv.head(10).plot(kind='bar', color='blue', alpha=0.7)
plt.title('Top 10 Actors in TV Shows', fontsize = 21)
plt.xlabel('Actor')
plt.ylabel('Count')
plt.show()

# Plotting the top 10 actors in movies
plt.figure(figsize=(14, 6))
actor_counts_movie.head(10).plot(kind='bar', color='green', alpha=0.7)
plt.title('Top 10 Actors in Movies', fontsize = 21)
plt.xlabel('Actor')
plt.ylabel('Count')
plt.show()
```

```python
# Filter the data where director_NA is not equal to flix['director']
filtered_data = flix[flix['director'] != 'director_NA']

# Extracting the directors columns
directors = filtered_data['director']

# Counting the occurrences of directors for each content type
director_counts_tv = directors[flix['type'] == 'TV Show'].value_counts()
director_counts_movie = directors[flix['type'] == 'Movie'].value_counts()

# Plotting the top 10 directors in TV shows
plt.figure(figsize=(14, 6))
director_counts_tv.head(10).plot(kind='bar', color='red', alpha=0.7)
plt.title('Top 10 Directors in TV Shows', fontsize = 21)
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()

# Plotting the top 10 directors in movies
plt.figure(figsize=(14, 6))
director_counts_movie.head(10).plot(kind='bar', color='orange', alpha=0.7)
plt.title('Top 10 Directors in Movies', fontsize = 21)
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
# Percentage distribution of content types in each country
country_content_distribution = flix.groupby('country')['type'].value_counts(normalize=True).mul(100).unstack().fillna(0)

# Plotting the percentage distribution of content types in each country
fig, ax = plt.subplots(figsize=(20, 6))
country_content_distribution.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Percentage Distribution of Content Types in Each Country', fontsize = 25)
ax.set_xlabel('Country')
ax.set_ylabel('Percentage')
ax.legend(title='Content Type')
ax.set_xticklabels(country_content_distribution.index, rotation=90)
plt.show()
```
